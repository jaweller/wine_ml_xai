```python
# -------------------------
# Ara Sınav Ödevi Notebook
# Klasik ML + PCA + LDA + XAI (SHAP)
# Veri seti: sklearn.datasets.load_wine (3 sınıf)
# -------------------------
!pip install -q scikit-learn pandas matplotlib seaborn xgboost shap
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, roc_curve, confusion_matrix
import xgboost as xgb
import shap
import warnings
warnings.filterwarnings("ignore")
sns.set(style='whitegrid')


# ---------- 1. Veri Setinin Yüklenmesi ----------
dataset_name = 'wine'
raw = datasets.load_wine()
X = pd.DataFrame(raw.data, columns=raw.feature_names)
y = pd.Series(raw.target, name='target')
print("Veri seti yüklenmiştir. Boyutlar:")
print("X shape:", X.shape)
print("y shape:", y.shape)

# İlk 5 satırı göster
display(X.head())
display(y.head())


# Özellikler X, hedef y olarak ayrıldı. Wine veri setinde 13 özellik, 3 sınıf var.

# ---------- 2. Veri Seti Kalite Kontrolleri ----------
# 2.1 Eksik Değer Analizi
missing_per_col = X.isnull().sum()
print("\nHer sütunda eksik değer sayısı:\n", missing_per_col)

# Eğer varsa doldurma mean ile
if missing_per_col.sum() > 0:
    X = X.fillna(X.mean())
    print("Eksik değerler mean ile dolduruldu.")

# 2.2 Aykırı Değer (Outlier) Analizi - IQR yöntemi ve boxplot
def detect_iqr_outliers(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return df[(df[col] < lower) | (df[col] > upper)]

outlier_summary = {}
for col in X.columns:
    outliers = detect_iqr_outliers(X, col)
    outlier_summary[col] = outliers.shape[0]
outlier_summary = pd.Series(outlier_summary)
print("\nHer sütundaki IQR aykırı değer sayısı (satır sayısı, toplam örnek sayısı ile karşılaştırın):")
display(outlier_summary.sort_values(ascending=False))

# Aykırı değerler ortalama/variance temelli modelleri etkileyebilir
# İhtiyaç halinde RobustScaler veya winsorize/trim uygulanabilir
# Robust scaler veya aykırı değerleri winsorize/trim ile ele almak gerekebilir.

# 2.3 Veri Tipi ve Dağılım İncelemesi
dtypes = X.dtypes
num_cols = X.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = X.select_dtypes(exclude=[np.number]).columns.tolist()

print("\nSayısal değişken sayısı:", len(num_cols))
print("Kategorik değişken sayısı:", len(cat_cols))
print("\nHer sütunun dtype bilgisi:")
display(dtypes)

# ---------- 3. Keşifsel Veri Analizi (EDA) ----------
# 3.1 İstatistiksel Özellikler
stats = X.describe().T
stats['median'] = X.median()
stats['q1'] = X.quantile(0.25)
stats['q3'] = X.quantile(0.75)
display(stats[['mean','median','min','max','std','q1','q3']])

# 3.2 Korelasyon Matrisi
corr = X.corr(method='pearson')
plt.figure(figsize=(12,10))
sns.heatmap(corr, annot=False, cmap='coolwarm', center=0)
plt.title('Pearson Korelasyon Matrisi (Isı haritası)')
plt.show()

# En yüksek korelasyonlu 3 çift
corr_unstack = corr.abs().where(np.triu(np.ones(corr.shape), k=1).astype(bool)).stack()
top3 = corr_unstack.sort_values(ascending=False).head(3)
print("\nEn yüksek korelasyonlu 3 çift (mutlak değer):")
display(top3)

# 3.3 Boxplot analizi - tüm özellikler için
plt.figure(figsize=(14,8))
X.boxplot(rot=90)
plt.title('Tüm özellikler için boxplot')
plt.tight_layout()
plt.show()

# Kısa yorum:
# - Bazı özelliklerde uç değerler (outlier) göze çarpıyor; gerektiğinde RobustScaler veya transform kullanmak mantıklı.

# ---------- 4. Veri Ölçeklendirme (Scaling) ----------
scaler = StandardScaler()  # önerilen
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
display(X_scaled.head())

# Kısa not: Ölçeklendirilmiş veri X_scaled olarak kaydedildi.

# ---------- 5. Veri Setinin Bölünmesi ----------
# Bölme oranları: %70 train, %10 val, %20 test
# Yöntem: önce test %20 ayır -> kalan %80'i train+val olarak al -> ondan train=%70/(%80)=0.875, val=%12.5 of original (=> 0.125)
X_temp, X_test, y_temp, y_test = train_test_split(X_scaled, y, test_size=0.20, stratify=y, random_state=42)
# X_temp'in %12.5/(80%) = 0.125/0.8 = 0.15625? Easier: split X_temp into train (87.5% of temp) and val (12.5% of temp)
X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size=0.125, stratify=y_temp, random_state=42)
print("Train/Val/Test shapes:", X_train.shape, X_val.shape, X_test.shape)

# ---------- 6. Özellik Seçimi ve Boyut İndirgeme ----------
# Üç veri temsili oluşturulacak: ham (X_scaled), PCA ile, LDA ile

# 6.1 PCA
pca_full = PCA()
pca_full.fit(X_train)
explained_ratio = pca_full.explained_variance_ratio_
# Bileşen sayısını: "explained variance ratio ortalamasından büyük olan" bileşenlerin sayısı
threshold = explained_ratio.mean()
n_components_pca = sum(explained_ratio > threshold)
if n_components_pca < 2:
    n_components_pca = 2  # en az 2 bileşen seçimi (2D scatter için)
print("PCA explained variance ratio mean:", threshold)
print("Seçilen PCA bileşen sayısı:", n_components_pca)

# Explained variance grafiği
plt.figure(figsize=(8,5))
plt.plot(np.cumsum(explained_ratio), marker='o')
plt.xlabel('n_components')
plt.ylabel('Cumulative explained variance')
plt.title('PCA Cumulative Explained Variance')
plt.grid(True)
plt.axhline(y=np.cumsum(explained_ratio)[n_components_pca-1], color='r', linestyle='--', label=f'selected {n_components_pca}')
plt.legend()
plt.show()

# PCA ile veriyi dönüştür
pca_transformer = PCA(n_components=n_components_pca)
X_train_pca = pd.DataFrame(pca_transformer.fit_transform(X_train),
                           columns=[f"PCA{i+1}" for i in range(n_components_pca)])
X_val_pca = pd.DataFrame(pca_transformer.transform(X_val),
                         columns=[f"PCA{i+1}" for i in range(n_components_pca)])
X_test_pca = pd.DataFrame(pca_transformer.transform(X_test),
                          columns=[f"PCA{i+1}" for i in range(n_components_pca)])

# 2D scatter plot (ilk iki PCA bileşeni)
plt.figure(figsize=(8,6))
for cls in np.unique(y_train):
    idx = (y_train.reset_index(drop=True) == cls)
    plt.scatter(
    X_train_pca.loc[idx, "PCA1"],
    X_train_pca.loc[idx, "PCA2"],
    label=f"class {cls}",
    alpha=0.7
)
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.title('Train set - İlk iki PCA bileşeni ile sınıfların ayrışması')
plt.legend()
plt.show()

# 6.2 LDA
# LDA için sınıf sayısından <= n_components <= n_classes-1 kuralı vardır. Wine için (3 sınıf) n_components<=2.
# Bu nedenle, n_components 2 olarak ayarlanmalıdır.
lda_model = LDA(n_components=min(X_train.shape[1], y_train.nunique() - 1))
X_train_lda = pd.DataFrame(lda_model.fit_transform(X_train, y_train))
X_val_lda = pd.DataFrame(lda_model.transform(X_val))
X_test_lda = pd.DataFrame(lda_model.transform(X_test))
print("\nLDA sonucu bileşen sayısı (sklearn kısıtından dolayı):", X_train_lda.shape[1])

# İlk iki LDA bileşeni ile görselleştirme
plt.figure(figsize=(8,6))
for cls in np.unique(y_train):
    idx = (y_train.reset_index(drop=True) == cls)
    plt.scatter(X_train_lda.loc[idx,0], X_train_lda.loc[idx,1], label=f'class {cls}', alpha=0.7)
plt.xlabel('LDA 1')
plt.ylabel('LDA 2')
plt.title('Train set - İlk iki LDA bileşeni ile sınıfların ayrışması')
plt.legend()
plt.show()

# Kısa yorum:
# PCA: bilgi kaybı olabilir
# LDA: sınıf ayrımını maksimize eder, max 2 bileşen (3 sınıf - 1)

# ---------- 7. Makine Öğrenmesi Modellerinin Kurulması ----------
# Modeller listesi
models = {
    "LogisticRegression": LogisticRegression(max_iter=1000, random_state=42),
    "DecisionTree": DecisionTreeClassifier(random_state=42),
    "RandomForest": RandomForestClassifier(n_estimators=100, random_state=42),
    "XGBoost": xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42),
    "GaussianNB": GaussianNB()
}

# Veri temsilleri
representations = {
    "raw": (X_train, X_val, X_test),
    "pca": (X_train_pca, X_val_pca, X_test_pca),
    "lda": (X_train_lda, X_val_lda, X_test_lda)
}

# Eğitim ve validation değerlendirme
from collections import defaultdict
val_results = []

def safe_roc_auc(y_true, y_score):
    # multi-class için ovo/ovr handled by sklearn
    try:
        if len(np.unique(y_true)) == 2:
            return roc_auc_score(y_true, y_score)
        else:
            return roc_auc_score(y_true, y_score, multi_class='ovr')
    except Exception:
        return np.nan

for rep_name, (Xtr, Xv, Xt) in representations.items():
    for model_name, model in models.items():
        # fit
        model.fit(Xtr, y_train)
        # prediction
        y_pred = model.predict(Xv)
        # for roc-auc, need scores / probas
        try:
            y_proba = model.predict_proba(Xv)
            # if multiclass, pass the probability matrix
            roc = safe_roc_auc(y_val, y_proba)
        except Exception:
            # try decision_function
            try:
                y_score = model.decision_function(Xv)
                roc = safe_roc_auc(y_val, y_score)
            except Exception:
                roc = np.nan
        acc = accuracy_score(y_val, y_pred)
        prec = precision_score(y_val, y_pred, average='weighted', zero_division=0)
        rec = recall_score(y_val, y_pred, average='weighted', zero_division=0)
        f1 = f1_score(y_val, y_pred, average='weighted', zero_division=0)
        val_results.append({
            "representation": rep_name,
            "model": model_name,
            "accuracy": acc,
            "precision": prec,
            "recall": rec,
            "f1": f1,
            "roc_auc": roc
        })
print("Modeller eğitildi ve validation üzerinde değerlendirildi.")

# ---------- 8. Validation Performanslarının Ölçülmesi ----------
val_df = pd.DataFrame(val_results)
# Tablo
display(val_df.sort_values(by='f1', ascending=False).reset_index(drop=True))

# Kısa yorum:
# - Bu tabloda validation sonuçları gösteriliyor. En iyi modeli seçerken f1/roc_auc kombinasyonuna bakılır.

# ---------- 9. En İyi Modelin Test Üzerinde Değerlendirilmesi ----------
# En iyi modeli validation'daki en yüksek f1 skoruna göre seçilir
best_row = val_df.sort_values(by='f1', ascending=False).iloc[0]
print("Validation'a göre en iyi model (f1):")
display(best_row)

best_rep = best_row['representation']
best_model_name = best_row['model']
print("Seçilen temsil:", best_rep, "Model:", best_model_name)

# Yeniden eğit: seçilen model ve temsildeki X_train tamamı üzerinde fit edip test setine bak
Xtr, Xv_unused, Xt = representations[best_rep]
# Fit model (yeniden initialize to avoid side-effects)
best_model = models[best_model_name]
best_model.fit(Xtr, y_train)
y_test_pred = best_model.predict(Xt)
# prob/score
try:
    y_test_proba = best_model.predict_proba(Xt)
    test_roc = safe_roc_auc(y_test, y_test_proba)
except Exception:
    try:
        y_test_score = best_model.decision_function(Xt)
        test_roc = safe_roc_auc(y_test, y_test_score)
    except Exception:
        y_test_proba = None
        test_roc = np.nan

test_acc = accuracy_score(y_test, y_test_pred)
test_prec = precision_score(y_test, y_test_pred, average='weighted', zero_division=0)
test_rec = recall_score(y_test, y_test_pred, average='weighted', zero_division=0)
test_f1 = f1_score(y_test, y_test_pred, average='weighted', zero_division=0)

print("\nTest set performans (seçilen model):")
print(f"Accuracy: {test_acc:.4f}")
print(f"Precision (weighted): {test_prec:.4f}")
print(f"Recall (weighted): {test_rec:.4f}")
print(f"F1-score (weighted): {test_f1:.4f}")
print(f"ROC-AUC: {test_roc:.4f}")

# 9.2 Confusion Matrix
cm = confusion_matrix(y_test, y_test_pred)
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title(f'Confusion Matrix - {best_model_name} on Test Set ({best_rep})')
plt.show()

# 9.3 ROC Eğrisi (multiclass için her sınıfın ROC eğrisi)
if y_test_proba is not None and y_test_proba.shape[1] > 1:
    # Binarize y_test for one-vs-rest plotting
    from sklearn.preprocessing import label_binarize
    n_classes = len(np.unique(y))
    y_test_bin = label_binarize(y_test, classes=np.arange(n_classes))
    plt.figure(figsize=(8,6))
    for i in range(n_classes):
        fpr, tpr, _ = roc_curve(y_test_bin[:,i], y_test_proba[:,i])
        auc = roc_auc_score(y_test_bin[:,i], y_test_proba[:,i])
        plt.plot(fpr, tpr, label=f'class {i} (AUC = {auc:.3f})')
    plt.plot([0,1],[0,1],'k--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curves (Test set)')
    plt.legend()
    plt.show()
else:
    # binary veya no proba
    if y_test_proba is not None:
        fpr, tpr, _ = roc_curve(y_test, y_test_proba[:,1])
        auc = roc_auc_score(y_test, y_test_proba[:,1])
        plt.figure(figsize=(6,5))
        plt.plot(fpr, tpr, label=f'AUC = {auc:.3f}')
        plt.plot([0,1],[0,1],'k--')
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('ROC Curve (Test set)')
        plt.legend()
        plt.show()

# Threshold etkisi:
# - ROC ve AUC bize sınıflandırıcı için genel ayrıştırma gücünü söyler. Ancak karar eşiği (threshold) seçimi precision/recall dengesini değiştirir.
# - Eğer belirli bir sınıf için false negative maliyeti yüksekse threshold düşürülerek recall yükseltilebilir, precision düşebilir.

# ---------- 10. XAI - SHAP Analizi ----------
# 10.1 En iyi validation model için SHAP
# Şu anki best_model değişkeni kullanılacak. SHAP explainer model tipine göre seçiliyor.

print("\nSHAP analizi başlatılıyor (seçili model may be tree or linear).")

# SHAP için testten ilk 200 örnek alındı (hesap maliyeti azaltmak için)
shap_sample = Xt.copy().reset_index(drop=True)
shap_sample_small = shap_sample.iloc[:200, :]  # yeterince küçük tut

# Explainer seçimi:
if isinstance(best_model, (RandomForestClassifier, DecisionTreeClassifier, xgb.XGBClassifier)):
    try:
        explainer = shap.TreeExplainer(best_model)
        shap_values = explainer.shap_values(shap_sample_small)
    except Exception as e:
        # fallback kernel
        explainer = shap.KernelExplainer(best_model.predict_proba, shap_sample_small.iloc[:50,:])
        shap_values = explainer.shap_values(shap_sample_small)
else:
    # logistic, nb -> KernelExplainer veya LinearExplainer
    try:
        explainer = shap.LinearExplainer(best_model, shap_sample_small, feature_dependence="independent")
        shap_values = explainer.shap_values(shap_sample_small)
    except Exception:
        explainer = shap.KernelExplainer(best_model.predict_proba, shap_sample_small.iloc[:50,:])
        shap_values = explainer.shap_values(shap_sample_small)

# summary_plot
plt.title("SHAP Summary Plot (bar)")
try:
    # For tree models shap_values may be list (one per class)
    if isinstance(shap_values, list):
        # show bar for class 0.. use mean(|shap|)
        for i, cls_shap in enumerate(shap_values):
            plt.figure()
            shap.summary_plot(cls_shap, shap_sample_small, plot_type="bar", show=False)
            plt.title(f"SHAP bar plot - class {i}")
            plt.show()
    else:
        shap.summary_plot(shap_values, shap_sample_small, plot_type="bar")
except Exception as e:
    print("SHAP summary plot hata:", e)

# summary_plot (dot)
try:
    if isinstance(shap_values, list):
        for i, cls_shap in enumerate(shap_values):
            plt.figure()
            shap.summary_plot(cls_shap, shap_sample_small, show=False)
            plt.title(f"SHAP summary (dot) - class {i}")
            plt.show()
    else:
        shap.summary_plot(shap_values, shap_sample_small)
except Exception as e:
    print("SHAP dot summary hata:", e)

# bar plot kullanıldı.
# Kısa yorum:
# - SHAP sıralaması modelin hangi feature'ları kararlarında daha çok kullandığını gösterir.
# - Eğer model tree-based ise TreeExplainer hızlı ve güvenilirdir. Logistic için LinearExplainer uygundur.
# 10.2 PCA ve LDA temsilleri için SHAP karşılaştırması
# Not: PCA/LDA bileşenleri özellik olduğundan, SHAP bileşenlerin etkisini gösterecek.
# PCA temsilinde best model tekrar eğit (eğer en iyi model PCA temsilde değilse yine de SHAP yapıyoruz)
# Burada PCA temsilinde aynı model istenirse aşağıdaki kodu kullanabilirsiniz:
if best_rep != 'pca':
    # modeli PCA verisi üzerinde de eğit ve SHAP uygula
    model_pca = models[best_model_name]
    model_pca.fit(X_train_pca, y_train)
    shap_sample_pca = X_test_pca.reset_index(drop=True).iloc[:200,:]
    try:
        explainer_pca = shap.TreeExplainer(model_pca) if isinstance(model_pca, (RandomForestClassifier, DecisionTreeClassifier, xgb.XGBClassifier)) else shap.LinearExplainer(model_pca, shap_sample_pca)
        shap_values_pca = explainer_pca.shap_values(shap_sample_pca)
        print("\nPCA temsili için SHAP (bar) gösteriliyor:")
        if isinstance(shap_values_pca, list):
            shap.summary_plot(shap_values_pca[0], shap_sample_pca, plot_type="bar")
        else:
            shap.summary_plot(shap_values_pca, shap_sample_pca, plot_type="bar")
    except Exception as e:
        print("PCA SHAP hata:", e)

# LDA temsili için SHAP
if best_rep != 'lda':
    model_lda = models[best_model_name]
    model_lda.fit(X_train_lda, y_train)
    shap_sample_lda = X_test_lda.reset_index(drop=True).iloc[:200,:]
    try:
        explainer_lda = shap.TreeExplainer(model_lda) if isinstance(model_lda, (RandomForestClassifier, DecisionTreeClassifier, xgb.XGBClassifier)) else shap.LinearExplainer(model_lda, shap_sample_lda)
        shap_values_lda = explainer_lda.shap_values(shap_sample_lda)
        print("\nLDA temsili için SHAP (bar) gösteriliyor:")
        if isinstance(shap_values_lda, list):
            shap.summary_plot(shap_values_lda[0], shap_sample_lda, plot_type="bar")
        else:
            shap.summary_plot(shap_values_lda, shap_sample_lda, plot_type="bar")
    except Exception as e:
        print("LDA SHAP hata:", e)

# Kısa yorum:
# - PCA bileşenleri lineer kombinasyon oldukları için SHAP'ta 'bileşenlerin' etkisini görürsünüz;
#   yorumlarken bileşenlerin hangi orijinal özelliklerle ilişkili olduğunu (loading'leri) kontrol etmek gerek.
# - LDA bileşenleri sınıf ayrımını maksimize ettiğinden, SHAP LDA bileşenlerinin sınıf ayrımına direkt etkisini gösterebilir.
# - SHAP sonuçları ile model performansı arasındaki ilişki: yüksek performanslı modeller genellikle (ama her zaman değil) daha tutarlı SHAP önem sıralamalarına sahip olur;
# önem sıralaması domain bilgisiyle tutarlı mı kontrol edin.

# ---------- Sonuçların kaydedilmesi ----------
# Validation karşılaştırma tablosu:
display(val_df.sort_values(by=['representation','f1'], ascending=[True, False]).reset_index(drop=True))

# Test performansı (seçilen model) çıktılarını bir dict olarak kaydet
test_perf = {
    "model": best_model_name,
    "representation": best_rep,
    "accuracy": test_acc,
    "precision": test_prec,
    "recall": test_rec,
    "f1": test_f1,
    "roc_auc": test_roc
}
print("\nTest performansı kaydedildi:", test_perf)
```
